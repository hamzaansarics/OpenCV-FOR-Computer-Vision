```python
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread('D:\\4.1.05.tiff')
img2=cv2.imread('D:\\4.1.06.tiff')
And=cv2.bitwise_and(img1,img2)
Or=cv2.bitwise_or(img1,img2)
Not=cv2.bitwise_not(img1)
Xor=cv2.bitwise_xor(img1,img2)
titles=['And','or','Not','Xor']
images=[And,Or,Not,Xor]
row=1
val=1
a=0
for s in images:
    plt.subplot(row,2,val)
    plt.title(titles[a])
    plt.imshow(images[a])
    val+=1
    a+=1
    if val==3 and row!=2:
        row+=1
        val=1
        plt.show()
    if val==3 and row==3:
        break
plt.show()
```
