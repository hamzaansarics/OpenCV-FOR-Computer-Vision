```python
import matplotlib.pyplot as plt
import cv2
img=cv2.imread('D:\\girl.jpg')

plt.subplot(1,3,1)
red=cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(red,color='r')

plt.subplot(1,3,2)
green=cv2.calcHist([img],[1],None,[256],[0,256])
plt.plot(green,color='g')
plt.subplot(1,3,3)
blue=cv2.calcHist([img],[2],None,[256],[0,256])
plt.plot(blue,color='b')
```
