```python
import cv2,random
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread(r'C:\Users\Hamza\Desktop/Gia.jpg')
rows,colums,channels=img.shape
new_img=np.zeros((img.shape),np.uint8)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
p=0.09
for r in range(rows):
    for c in range(colums):
        rnd=random.random()
        if rnd<p/2:
            new_img[r][c]=[0,0,0]
        elif rnd<p:
            new_img[r][c]=[255,255,255]
        else:
            new_img[r][c]=img[r][c]
denoised=cv2.medianBlur(new_img,3)
images=[img,new_img,denoised]
titles=['Orignal','Noised Image','Denoised Image']
col=1
for v in range(len(images)):
    plt.subplot(1,3,col)
    plt.imshow(images[v])
    plt.title(titles[v])
    plt.xticks([])
    plt.yticks([])
    col+=1
plt.show()
```
