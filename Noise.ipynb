```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
img=cv2.imread('E://TT.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rows,colums,channels=img.shape
noised_image=np.zeros(img.shape,np.uint8)
p=0.2
l=0
for r in range(rows):
    for c in range(colums):
        l+=100
        
        rndm=random.random()
        if rndm<p/2:
            #salt noise
            noised_image[r][c]=[255,255,255]
        elif rndm<p:
            #peppor noise
            noised_image[r][c]=[0,0,0]         
        else:
            noised_image[r][c]=img[r][c]
              
plt.imshow(noised_image)
plt.show()
denoised=cv2.medianBlur(noised_image,5)
plt.imshow(denoised)
plt.show()
```
