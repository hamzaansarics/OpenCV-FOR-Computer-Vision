```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('D:\\Ansari.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
R,G,B=cv2.split(img)
hist,binz=np.histogram(img.ravel(),256,[0,256])
plt.plot(hist)
plt.show()
Red,biz=np.histogram(R.ravel(),256,[0,256])
plt.plot(Red)
plt.show()
Green,biz=np.histogram(G.ravel(),256,[0,256])
plt.plot(Green)
plt.show()
Blue,biz=np.histogram(B.ravel(),256,[0,256])
plt.plot(Blue)
plt.show()
```
