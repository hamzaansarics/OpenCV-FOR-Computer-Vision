```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('D:\\4.1.05.tiff')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
rows,coloums,channels=img.shape
shifting=np.float32(([1,0,-50],[0,1,150]))
shifted_image=cv2.warpAffine(img,shifting,(rows,coloums))
plt.imshow(shifted_image)
plt.show()
```
