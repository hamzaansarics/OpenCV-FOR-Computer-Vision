```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('D:\\Ansari.jpg',1)
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh=cv2.threshold(gray,80,255,0)
plt.imshow(thresh,cmap='gray')
plt.show()
img1,contours,hirecy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
outptu=cv2.drawContours(img,contours,-1,(255,0,0),2)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
