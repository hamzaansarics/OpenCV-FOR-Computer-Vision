```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('D:\\girl.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
R,G,B=cv2.split(img)

Red=cv2.equalizeHist(R)
Green=cv2.equalizeHist(G)
Blue=cv2.equalizeHist(B)
Adjusted=cv2.merge((Red,Green,Blue))

# clahi=cv2.createCLAHE()
clahi=cv2.createCLAHE(clipLimit=1.0,tileGridSize=(8,8))
Red_C=clahi.apply(R)
Green_C=clahi.apply(G)
Blue_C=clahi.apply(B)
CO=cv2.merge((Red_C,Green_C,Blue_C))


Red_=cv2.equalizeHist(Red_C)
Green_=cv2.equalizeHist(Red_C)
Blue_=cv2.equalizeHist(Blue_C)
CLA=cv2.merge((Red_,Green_,Blue_))

images=[img,Adjusted,CO,CLA]
titles=['Orignal Image','Simple Adusted','Clahe','Clahi Adjusted']

row=1
col=1
for i in range(len(images)):
    plt.subplot(row,3,col)
    plt.imshow(images[i])
    plt.title(titles[i])
    col+=1
    if col==4:
        row+=1
        col=1
        plt.show()
plt.show()
```
