```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('D:\\girl.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
box_blur=cv2.boxFilter(img,-1,(21,21))
blur=cv2.blur(img,(21,21))
gabl=cv2.GaussianBlur(img,(21,21),0)
plt.subplot(1,3,1)
plt.imshow(box_blur)
plt.subplot(1,3,2)
plt.imshow(blur)
plt.subplot(1,3,3)
plt.imshow(gabl)
plt.show()
```
