```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('D:\\7.1.01.tiff',0)
#-------------Operations on Grayscale images------------------#
th=0
max_val=255
#making Kernel
ker=np.ones((5,5),np.uint8)
#Thresholding
ret,bi=cv2.threshold(img,th,max_val,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
#Erosion Operation
erosion=cv2.erode(bi,ker,iterations=3)
plt.imshow(erosion,cmap='gray')
plt.show()
#Dilatoins Operation
dilation=cv2.dilate(bi,ker,iterations=3)
plt.imshow(dilation,cmap='gray')
plt.show()
#Gradient Operations
gradient=cv2.morphologyEx(bi,cv2.MORPH_GRADIENT,ker)
plt.imshow(gradient,cmap='gray')
plt.show()
#-------------Operations on Color images------------------#
img1=cv2.imread('D:\\4.2.03.tiff',1)
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.show()
#Erosion Color
erosion_color=cv2.erode(img1,ker,iterations=3)
plt.imshow(erosion_color)
plt.show()
#Dilation Color
dilation_color=cv2.dilate(img1,ker,iterations=3)
plt.imshow(dilation_color)
plt.show()

#Gradint Color
gradient=cv2.morphologyEx(img1,cv2.MORPH_GRADIENT,ker)
plt.imshow(gradient)
plt.show()
```
