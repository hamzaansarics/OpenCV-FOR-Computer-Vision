```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('D:\\4.1.06.tiff')
rows,colums,channels=img.shape
point1=np.float32([[0,0],[400,0],[0,400],[400,400]])
point2=np.float32([[0,0],[500,0],[0,500],[500,500]])
# point3=np.float32([[400,200],[150,200],[200,350]])
A=cv2.getPerspectiveTransform(point1,point2)
output=cv2.warpPerspective(img,A,(400,400))
plt.imshow(output)
plt.show()
```
