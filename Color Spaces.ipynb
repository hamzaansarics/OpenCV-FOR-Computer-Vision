```python
import cv2
img=cv2.imread('D:\\girl.jpg',1)
#color_space -->[COLOR_BGR2RGB]
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.imshow('COol',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

#Colors counter and showing colors spaces in cv2
count=0
for s in dir(cv2):
    if s.startswith('COLOR_'):
        count+=1
        print(s)
print(count)
```
