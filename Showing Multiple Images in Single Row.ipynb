```python
import cv2
import matplotlib.pyplot as plt
img1=cv2.imread('D:\\girl.jpg')
img2=cv2.imread('D:\\leena.tiff')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
images=[img1,img2]
title=['American Girl','Leena']
a=0
val=1
for s in range(len(images)):
        plt.subplot(1,2,val)
        plt.imshow(images[s])
        plt.title(title[a])
        a+=1
        val+=1
plt.show()
```
