```python
import matplotlib.pyplot as plt
import cv2
img=cv2.imread('D:\\girl.jpg',1)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
R,G,B=cv2.split(img)
plt.subplot(1,4,1)
plt.imshow(img)
plt.title('Orignal Image')
plt.subplot(1,4,2)
plt.hist(R.ravel(),256,[0,256])
plt.title('Red ')
plt.subplot(1,4,3)
plt.hist(G.ravel(),256,[0,256])
plt.title('Green')
plt.subplot(1,4,4)
plt.hist(B.ravel(),256,[0,256])
plt.title('Blue ')
plt.show()
```
