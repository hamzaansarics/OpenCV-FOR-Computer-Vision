```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('D:\\4.1.06.tiff')
rows,colums,channels=img.shape
point1=np.float32([[100,100],[300,100],[200,300]])
point2=np.float32([[100,250],[200,100],[300,300]])
# point3=np.float32([[400,200],[150,200],[200,350]])
A=cv2.getAffineTransform(point1,point2)
output=cv2.warpAffine(img,A,(rows,colums))
plt.imshow(output)
plt.show()
```
